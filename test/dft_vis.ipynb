```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pltx = []
plty = []

fx = []
fy = []
```

```python
with open("../data/square.csv", "r") as f:
    del plty[:]
    
    i = 0
    for line in f:
        i += 1
        if i < 3:
            continue
            
        line = line.split(",")
        plty.append(float(line[1]))
        
    plt.figure()
    plt.plot(plty)
    plt.show()
```

```python
with open("dft.txt", "r") as f:
    del fy[:]
    
    for line in f:
        line = line.split(",")
        fy.append(float(line[1]))
        
    plt.figure()
    plt.yscale('log')
    plt.plot(fy[:len(fy)//2])
    plt.show()
```
